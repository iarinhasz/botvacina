Isso não nos ajuda muito, então vamos categorizar os sintomas 

```python
categorias_sintomas = {
    'Sintomas Comuns de Gripe': ['Tosse', 'Coriza', 'Febre', 'Dor de Garganta'],
    'Sintomas Respiratórios': ['Dispneia', 'Dificuldade para respirar', 'Falta de Ar'],
    'Sintomas Gastrointestinais': ['Diarreia', 'Vômito', 'Náusea','Dor Abdominal'],
    'Sintomas Neurológicos': ['Dor de Cabeça', 'Tontura', 'Confusão'],
    'Outros Sintomas': ['Fadiga', 'Mal-estar', 'Mialgia']
}
```

```python
def categorizar_sintomas(sintomas, outros):
    # Junta os dois campos e transforma em lista de sintomas individuais
    todos_sintomas = []
    
    if pd.notna(sintomas):
        todos_sintomas += [s.strip().upper() for s in sintomas.split(',')]
    
    if pd.notna(outros):
        todos_sintomas += [s.strip().upper() for s in outros.split(',')]
    
    # Categorias encontradas
    categorias_encontradas = set()

    # Procura cada sintoma em cada categoria
    for categoria, lista_sintomas in categorias_sintomas.items():
        for sintoma in lista_sintomas:
            if sintoma.upper() in todos_sintomas:
                categorias_encontradas.add(categoria)
    
    return list(categorias_encontradas)
```

```python
df['categorias'] = df.apply(lambda row: categorizar_sintomas(row['sintomas'], row['outrosSintomas']), axis=1)
print(df[['sintomas', 'outrosSintomas', 'categorias']].head())
```

```python
def responder(pergunta):
    pergunta = pergunta.lower()
    if "estado" in pergunta:
        estado = input("Digite o estado(nome ou sigla)")
        #quantas pessoas foram vacinadas no estado
        total = df[df['estado']== estado].shape[0]
        return f"Total de registros para o estado {estado}:{total}"

    elif "sexo" in pergunta:
        sexo = input("Digite o sexo (Feminino ou Masculino): ")
        total = df[df['sexo'].str.lower() == sexo].shape[0]
        return f"Total de vacinados do sexo {sexo}: {total}"

    elif "idade" in pergunta:
        idade = int(input("Digite a idade: "))
        total = df[df['idade'] == idade].shape[0]
        return f"Total de vacinados com {idade} anos: {total}"
    
    elif "raça" in pergunta or "cor" in pergunta:
        raca = input("Digite a raça/cor: ")
        total = df[df['racaCor'] == raca].shape[0]
        return f"Total de vacinados da raça/cor {raca}: {total}"
    
    elif "ano" in pergunta:
        ano = input("Digite o ano desejado (ex: 2021): ")
        df['ano'] = pd.to_datetime(df['data1dose']).dt.year
        total = df[df['ano'] == int(ano)].shape[0]
        return f"Total de vacinados no ano {ano}: {total}"

    else:
        return "Desculpe, não entendi a pergunta."

# Teste no notebook:
pergunta = input("Digite sua pergunta sobre vacinação(estado, sexo, idade, raça ou ano): ")
resposta = responder(pergunta)
print(resposta)
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
estado_dropdown = widgets.Dropdown(
    options=df['estado'].unique(),
    description='Estado:'
)
```

```python
sexo_dropdown = widgets.Dropdown(
    options=df['sexo'].unique(),
    description='Sexo:'
)
```

```python
dose_dropdown = widgets.Dropdown(
    options=[
        ('1ª Dose', 'dataPrimeiraDose'),
        ('2ª Dose', 'dataSegundaDose'),
        ('Todas as Doses', 'todas')
    ],
    description='Dose:'
)
df['dataPrimeiraDose'] = pd.to_datetime(df['dataPrimeiraDose'], errors='coerce')
df['dataSegundaDose'] = pd.to_datetime(df['dataSegundaDose'], errors='coerce')
```

```python
botao = widgets.Button(description="Buscar total")

output = widgets.Output()

def on_click(b):
    estado = estado_dropdown.value
    sexo = sexo_dropdown.value
    dose_escolhida = dose_dropdown.value
    
    filtro = (df['estado'] == estado) & (df['sexo'] == sexo)
    
    if dose_escolhida != 'todas':
        filtro = filtro & (df[dose_escolhida].notna())
    else:
        filtro = filtro & (
            df['dataPrimeiraDose'].notna() |
            df['dataSegundaDose'].notna()
        )
    
    total = df[filtro].shape[0]

    
    
    total = df[(df['estado'] == estado) & (df['sexo'] == sexo)].shape[0]
    
    output.clear_output()
    with output:
        print(f"Total de vacinados do sexo {sexo} no estado {estado} para {dose_escolhida}: {total}")

botao.on_click(on_click)

display(estado_dropdown, sexo_dropdown, dose_dropdown, botao, output)
```
